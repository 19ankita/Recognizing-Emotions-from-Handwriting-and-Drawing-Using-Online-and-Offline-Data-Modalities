```python
# Install Ghostscript
!apt-get install -y ghostscript

import os
import glob

input_dir = "results"   # folder where your SHAP pdfs are saved
output_dir = "results/compressed"
os.makedirs(output_dir, exist_ok=True)

# Compress all PDF files
for pdf in glob.glob(os.path.join(input_dir, "*.pdf")):
    base = os.path.basename(pdf)
    output_pdf = os.path.join(output_dir, base.replace(".pdf", "_small.pdf"))

    cmd = f"gs -sDEVICE=pdfwrite -dCompatibilityLevel=1.4 -dPDFSETTINGS=/ebook " \
          f"-dNOPAUSE -dQUIET -dBATCH -sOutputFile='{output_pdf}' '{pdf}'"

    os.system(cmd)
    print(f"Compressed: {base} -> {output_pdf}")

```
